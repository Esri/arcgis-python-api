Multi GPU training is only supported for Linux machine with multiple GPUs.

```python
from arcgis.gis import GIS
```

```python
gis = GIS('https://geosaurus.maps.arcgis.com', 'arcgis_python', 'P@ssword123')
```

```python
script_item = gis.content.get('afd1c9a88a6f4f04896b4172c0f3a78c')
```

```python
script_item.name
```

```python
filepath = script_item.download(file_name=script_item.name)
```

```python
import zipfile
import os
from pathlib import Path
with zipfile.ZipFile(filepath, 'r') as zip_ref:
    zip_ref.extractall(Path(filepath).parent)
```

```python
script_path = Path(os.path.join(os.path.splitext(filepath)[0]) + '.py')
```

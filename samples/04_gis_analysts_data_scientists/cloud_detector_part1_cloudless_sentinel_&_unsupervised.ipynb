## Introduction <a class="anchor" id="1"></a>

This notebook elaborates on the different ways cloud mask could be generated from satellite imagery and here sentinel imagery is used as the sample data.

It is a well-known fact that cloud presence causes problems for remote sensing analysis of surface properties like land use land cover classification, image compositing, change detection etc. In case of single scene image processing though its comparatively easier to filter out clouds, but for studies that uses large number of images, these would require automated approach for removing or masking out cloud affected areas.

Accordingly, here three methods are used for the cloud mask extraction:

   - First the sentinel2 cloudless python package is used, which is sentinel's hub cloud detector, and which works only on sentinel imagery.
    
   - Second an unsupervised model using the mlmodel framework is applied to get cloud mask. This can run on any kind of imagery.
    
   - Third a pixel classifier model based on Unet is trained.

The first two methods are elaborated in this notebook, and the third method will be described in the second part of the notebook, followed by comparison of the results of all the three approaches.

## Method 1: Cloud mask using Sentinel Cloudless <a class="anchor" id="4a"></a>

### Data Preparation <a class="anchor" id="5"></a>

#### Define Area of Interest on NYC with clouds <a class="anchor" id="6"></a>
The area of interest is defined using the four latitude and longitude values from a certain part of the NYC with presence of clouds, as can be seen from the images.

```python
#  extent in 3857 for NYU rain clouds
NYU_cloud_extent = {
    "xmin": -8231176.77,
    "ymin": 4967559.25,
    "xmax": -8242898.16,
    "ymax": 4973524.61,
    "spatialReference": {"wkid": 3857}
}
```

```python
# NYU - The respective scene having the above area is selected
selected = sentinel.filter_by(where="(Category = 1) AND (cloudcover >=0.05)",                              
                              geometry=arcgis.geometry.filters.intersects(NYU_cloud_extent))
df = selected.query(out_fields="AcquisitionDate, GroupName, CloudCover, DayOfYear",
                   order_by_fields="AcquisitionDate").sdf

df['AcquisitionDate'] = pd.to_datetime(df['acquisitiondate'], unit='ms')
df
```

#### Visualize Area of Interest on NYC with clouds <a class="anchor" id="7"></a>

```python
# The scene is selected with the least cloud cover and extracted using the selected NYC extent
amazon_clouds_scene4 = sentinel.filter_by('OBJECTID=352211')
amazon_clouds_scene4.extent = NYU_cloud_extent
amazon_clouds_scene4
```

#### Convert scene to arrray <a class="anchor" id="8"></a>

```python
# convert the scene to a numpy array
test_array = amazon_clouds_scene4.export_image(f='numpy_array')
```

```python
test_array.shape
```

```python
# Change array to float data type
test_array_sr = test_array/10000
test_array_sr = test_array_sr.astype('float32')
```

### Initiate cloud detector <a class="anchor" id="9"></a>
Finally the cloud detector is initiated using selected paramters. The values can be adjusted for different amount of cloud cover.

```python
# Initiate sentinel cloud detector 
cloud_detector = S2PixelCloudDetector(
    threshold=0.7,
    average_over=4,
    dilation_size=2,
    all_bands=True
)
```

### Generate cloud mask <a class="anchor" id="10"></a>

The initiated cloud detector is now used to extract the cloud mask from the selected scene

```python
%%time
test_cloud_mask = cloud_detector.get_cloud_masks(test_array_sr)
```

## Method 2: Cloud mask using unsupervised learning <a class="anchor" id="12"></a>

In this section the unsupervised learning method of k-means clustering is used detect clouds from imagery. First a scene from a preselected area of interest is accessed for detecting clouds.

### Accessing & Visualizing datasets  <a class="anchor" id="13"></a>

```python
iowa_test = Raster(r"D:\my\2021\10-October\cloud mask detector\data\iowacity1\rgbnir_iowa.tif")
```

```python
iowa_test.extent
```

```python
iowa_test.read().shape
```

```python
iowa_test.export_image(size=[500,501])
```

In the above scene, there is only clouds present in the area of interest which are to be identified using this method. 

### Data Pre-processing <a class="anchor" id="14"></a>

```python
iowa_test.name
```

Here the imagery name is 'rgbnir_iowa.tif', following which the name of the respective 4 bands of blue, green, red, and near infrared bands would be 'rgbnir_iowa.tif','rgbnir_iowa.tif_1', 'rgbnir_iowa.tif_2', 'rgbnir_iowa.tif_3'. These bands are to be used for defining the preprocessors.  

```python
iowa_test.band_names
```

This is a four band imagery so the preprocessors are first defined for scaling it.

```python
preprocessors = [('rgbnir_iowa.tif', 'rgbnir_iowa.tif_1', 'rgbnir_iowa.tif_2', 'rgbnir_iowa.tif_3', MinMaxScaler())]
```

```python
data = prepare_tabulardata(explanatory_rasters=[iowa_test], preprocessors=preprocessors)
```

```python
data.show_batch()
```

### Model Initialization <a class="anchor" id="15"></a>

Once the data is prepared an unsupervised model of k-means clustering from scikit-learn is used here for clustering the pixels into clouds and areas without clouds. Here the clustering model is passed inside MLModel as follows, with number of clusters as two in the parameters.

```python
model = MLModel(data, 'sklearn.cluster.KMeans', n_clusters=2, init='k-means++', random_state=43)
```

### Model Training <a class="anchor" id="16"></a>
Now the model is ready to be trained which would label the pixels into two classes of no clouds, and dense clouds. 

```python
model.fit()
```

```python
model.show_results()
```

### Cloud Prediction<a class="anchor" id="17"></a>

```python
pred_new = model.predict(explanatory_rasters=[iowa_test], 
                         prediction_type='raster',
                         output_raster_path=r"D:\my\2021\10-October\cloud mask detector\result\result2.tif")
```
